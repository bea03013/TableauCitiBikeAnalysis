```python
import pandas as pd
```

```python
bike = pd.read_csv("Raw/201306-citibike-tripdata.csv")
bike.head(3)
```

```python
bike = bike[["tripduration", "start station id", "start station name", "start station latitude", "start station longitude",
            "end station id", "end station name", "end station latitude", "end station longitude", "bikeid", "usertype",
            "birth year", "gender"]]
bike.head(3)
```

```python
bike = bike.rename(columns={"tripduration":"Trip Duration", "start station id":"S Station ID", "start station name":"S Station Name", 
                    "start station latitude":"S Station Lat", "start station longitude":"S Station Long", 
                    "end station id":"E Station ID", "end station name":"E Station Name", 
                    "end station latitude":"E Station Lat", "end station longitude":"E Station Long", 
                    "bikeid":"Bike ID", "usertype":"User", "birth year":"Birth Year", "gender":"Gender"})
bike.head(3)
```

```python
bike.to_csv("Data/station.csv", index=False, header=True)
```
